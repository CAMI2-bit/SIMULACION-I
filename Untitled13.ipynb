```python

import random as rd
import numpy as np
import matplotlib.pyplot as plt

a = 1
b = 3
c = 5

# Definimos la función f(x) de la distribución triangular
def f(x):
    if a <= x <= b:
        return (2*(x - a)) / ((c - a)*(b - a))
    elif b < x <= c:
        return (2*(c - x)) / ((c - a)*(c - b))
    else:
        return 0

# Valor máximo de f(x), necesario para el método de aceptación-rechazo
M = 2 / (c - a)

# Método de aceptación-rechazo
def aceptacion(n):
    lista = []
    while len(lista) < n:
        R1 = rd.random()
        R2 = rd.random()
        x = a + (c - a) * R1  # Generamos x entre [a, c]
        if R2 <= f(x) / M:
            lista.append(x)
    return lista


c = aceptacion(50000)

# Imprimimos una pequeña parte
print(c[:10])  # Solo imprimimos los primeros 10 para no saturar


x = np.linspace(a, c, 500)


plt.hist(c, bins=50, density=True, alpha=0.6, label='Histograma muestras')

plt.show()
```
