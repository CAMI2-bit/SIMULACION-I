 **Uniforme** $X \sim U(0,1)$

Funcion de densidad

\begin{equation}
f(x)=
\begin{cases}
\frac{1}{b-a} & \text{si } x \in (a, b) \\
0 & \text{si } x \notin (a, b)
\end{cases}
\end{equation}

image.png

Funcion acumulativa

\begin{equation}
F(x)=
\begin{cases}
0 & \text{si } x \leq a  \\
\frac{x - a}{b - a} & \text{si } x \in (a, b) \\
1 & \text{si } x \geq b
\end{cases}
\end{equation}


Image13.gif

Media

\begin{equation}
\mu = \frac{a+b}{2}, \quad a<x<b
\end{equation}

Varianza

\begin{equation}
\sigma^2 = \frac{(b-a)^2}{12}, \quad a<x<b
\end{equation}

Desviacion estandar

\begin{equation}
\sigma = \frac{(b-a)}{\sqrt{12}}, \quad a<x<b
\end{equation}

**Pseudocodigo**



1.   Generar $U \sim U(0,1)$
2.   Definir $x=a+(b-a)U$
3.   Retornar las x`s


```python
#Librerías
import numpy as np
import matplotlib.pyplot as plt
from random import random

#Función de la transformada inversa para la distribucion Uniforme
def uniforme(n):
  a,b=1,3 #Parametros
  x = [] #Lista vacia
  for i in range(n):
    R = random() #Numeros Aleatorios
    x.append(a+(b-a)*R)
  return x

#Muestras
A = uniforme(10000)

#Histograma
plt.hist(A, bins=100, density=True)
plt.title("Distribución Uniforme")


#Media y varianza
print("Media:", np.mean(A))
print("Varianza:", np.var(A))
print("Desviación estándar:", np.std(A))
```

 **Triangular** $X \sim U(0,1)$

Funcion de densidad
\begin{equation}
f(x) =
\begin{cases}
\displaystyle \frac{2(x - a)}{(b - a)(c - a)} & \text{para } a \leq x \leq c \\[10pt]
\displaystyle \frac{2(b - x)}{(b - a)(b - c)} & \text{para } c < x \leq b
\end{cases}
\end{equation}

image.png

Funcion Acumulativa
\begin{equation}
F(x) =
\begin{cases}
\displaystyle \frac{(x - a)^2}{(b - a)(c - a)} & \text{para } a \leq x \leq c \\[10pt]
\displaystyle 1 - \frac{(b - x)^2}{(b - a)(b - c)} & \text{para } c < x \leq b
\end{cases}
\end{equation}

image.png

Media

\begin{equation}
\mu = \frac{a+b+c}{3}, \quad a<x<b
\end{equation}

Varianza

\begin{equation}
\sigma^2 = \frac{a^2+b^2+c^2-ab-ac-bc}{18}, \quad a<x<b
\end{equation}

Desviacion estandar

\begin{equation}
\sigma = \sqrt{\frac{a^2+b^2+c^2-ab-ac-bc}{18}}, \quad a<x<b
\end{equation}

**Pseudocodigo**



1.   Generar $U \sim U(0,1)$
2.   Definir $$ x = \begin{cases}
a + \sqrt{R \cdot (b - a)(c - a)} & \text{si } R < \frac{b-a}{c-a} \\
c - \sqrt{(1 - R) \cdot (c - b)(c - a)} & \text{si } R \geq \frac{b-a}{c-a}
\end{cases} $$
3.   Retornar las x`s


```python
#Librerías
import numpy as np
import matplotlib.pyplot as plt
from random import random

#Función de la transformada inversa para la distribucion triangular
def triangular(n):
  a,b,c=1,3,5 #Parametros
  x = [] #Lista vacia
  for i in range(n):
    R = random() #Numeros Aleatorios
    if R<(b-a)/(c-a):
      x.append(a + np.sqrt(R * (b - a) * (c - a)))
    elif R>=(b-a)/(c-a):
      x.append(c - np.sqrt((1 - R) * (c - b) * (c - a)))
  return x

#Muestras
A = triangular(10000)

#Histograma
plt.hist(A, bins=100, density=True)
plt.title("Distribución Triangular")

#Media y varianza
print("Media:", np.mean(A))
print("Varianza:", np.var(A))
print("Desviación estándar:", np.std(A))
```

**Pseudocidigo**

1.   Generar $U \sim U(0,1)$

3.   Retornar las x`s



```python
import numpy as np
import matplotlib.pyplot as plt
from random import random

def exponencial(lam, n):
  x = []
  for i in range(n):
    R = random()
    x.append(-np.log(R)/lam)
  return x

exp = exponencial(1.5,100000)

plt.hist(exp, bins=100)

# Resultados teóricos y empíricos
lam = 1.5
media_teorica = 1 / lam
varianza_teorica = 1 / (lam** 2)
std_teorica = np.sqrt(varianza_teorica)

print("\n--- Resultados teóricos ---")
print(f"Media teórica: {media_teorica:.4f}")
print(f"Varianza teórica: {varianza_teorica:.4f}")
print(f"Desviación estándar teórica: {std_teorica:.4f}")

print("\n--- Resultados empíricos ---")
print(f"Media empirica: {np.mean(exp):.4f}")
print(f"varianza empirica: {np.var(exp):.4f}")
print(f"Desviacion estandar empirica:{np.std(exp):.4f}" )
```

 **m-Erlang** $X \sim Erlang(n,\lambda)$

Funcion de densidad

\begin{equation}
 f(n, \lambda) = \frac{\lambda (\lambda x)^{n-1} e^{-\lambda x}}{(n-1)!}, \quad x \geq 0
\end{equation}

image.png

Funcion de distribucion

\begin{equation}
F(x) = 1 - \sum_{k=0}^{n-1} \frac{(\lambda x)^k e^{-\lambda x}}{k!}, \quad x \geq 0
\end{equation}

image.png

Media

\begin{equation}
\mu =  \frac{n}{\lambda}
\end{equation}

Varianza
\begin{equation}
\sigma^2 =  \frac{n}{\lambda^2}
\end{equation}

Desviacion estandar
\begin{equation}
\sigma =  \frac{\sqrt{n}}{\lambda}
\end{equation}

**Pseudocidigo**

1.   Generar $U_1, U_2, ..., U_m \quad$ con $\quad U_i \sim U(0,1)$
2.   Definir $$ x = -\frac{1}{\lambda} \ln\left( \prod_{i=1}^m U_i \right) $$
3.   Retornar las x`s



```python
#Librerías
import numpy as np
import matplotlib.pyplot as plt
from random import random

def erlang(m, lam, n):
    x = []
    for _ in range(n):
        u = [random() for _ in range(m)]  #Generar m números aleatorios uniformes
        P = np.prod(u)                    #Calcular el producto de los U_i
        x.append(-np.log(P) / lam)        #Aplicar la transformada inversa
    return x

#Parámetros
m = 3       #Número de etapas (eventos)
lam = 2     #Tasa (parámetro de escala)
n = 10000   #Número de muestras

#Generar muestras
A = erlang(m, lam, n)

#Histograma
plt.hist(A, bins=100, density=True, alpha=0.7)
plt.title(f"Distribución de m-Erlang")
plt.xlabel("Valor")
plt.ylabel("Densidad")
plt.show()

#Media, varianza y desviación estándar teóricas
media_teorica = m / lam
varianza_teorica = m / (lam ** 2)
desviacion_teorica = np.sqrt(varianza_teorica)

#Resultados
print("\nResultados teóricos")
print(f"Media teórica: {media_teorica:.4f}")
print(f"Varianza teórica: {varianza_teorica:.4f}")
print(f"Desviación estándar teórica: {desviacion_teorica:.4f}")

print("\nResultados numericos")
print(f"Media muestral: {np.mean(A):.4f}")
print(f"Varianza muestral: {np.var(A):.4f}")
print(f"Desviación estándar muestral: {np.std(A):.4f}")
```

 **Gamma** $X \sim U(\alpha, \beta)$

 Función de densidad
\begin{equation}
f(x; \alpha, \beta) =
\begin{cases}
\frac{\beta^\alpha x^{\alpha - 1} e^{-\beta x}}{\Gamma(\alpha)} & \text{si } x > 0 \\
0 & \text{si } x \leq 0
\end{cases}
\end{equation}

image.png

Funcion Acumulativa
\begin{equation}
F(x; \alpha, \beta) = \frac{\gamma(\alpha, \beta x)}{\Gamma(\alpha)}
\end{equation}

Media

\begin{equation}
\mu =  \frac{\alpha}{\beta}
\end{equation}

Varianza
\begin{equation}
\sigma^2 =  \frac{\alpha}{\beta^2}
\end{equation}

Desviacion estandar
\begin{equation}
\sigma =  \frac{\sqrt{\alpha}}{\beta}
\end{equation}

**Pseudocidigo**

*CASO $\alpha=1$ (Algoritmo GS de Ahrens-Dieter)*

1.   Generar $U \sim U(0,1)$
2.   Definir $$ x = \frac{-In(U)}{\beta} $$
3.   Retornar x

*CASO $0<\alpha<1$ (EXPONENCIAL)*


```python
import numpy as np
from math import gamma, exp, log, sqrt
from scipy.stats import gamma as gamma_dist
import matplotlib.pyplot as plt

def gamma_ahrens_dieter(alpha, n_samples):
    """
    Implementación del algoritmo GS de Ahrens-Dieter para generar variables Gamma(α,1)
    con 0 < α < 1.

    Parámetros:
        alpha: parámetro de forma (0 < alpha < 1)
        n_samples: número de muestras a generar

    Devuelve:
        array de muestras Gamma(α,1)
    """
    samples = []
    b = (np.e + alpha)/np.e  # Constante b

    while len(samples) < n_samples:
        U1 = np.random.uniform()
        P = b * U1

        if P <= 1:
            # Paso 2 del algoritmo
            Y = P ** (1/alpha)
            U2 = np.random.uniform()
            if U2 <= exp(-Y):
                samples.append(Y)
        else:
            # Paso 3 del algoritmo
            Y = -log((b - P)/alpha)
            U2 = np.random.uniform()
            if U2 <= Y ** (alpha - 1):
                samples.append(Y)

    return np.array(samples)

# Implementación del Algoritmo GB de Cheng para Gamma(α,1) con α > 1

def gamma_cheng(alpha, n_samples):
    """
    Implementación del algoritmo GB de Cheng para generar variables Gamma(α,1)
    con α > 1.

    Parámetros:
        alpha: parámetro de forma (alpha > 1)
        n_samples: número de muestras a generar

    Devuelve:
        array de muestras Gamma(α,1)
    """
    samples = []
    a = 1/sqrt(2*alpha - 1)
    b = alpha - log(4)
    q = alpha + 1/a
    theta = 4.5
    d = 1 + log(theta)

    while len(samples) < n_samples:
        U1, U2 = np.random.uniform(size=2)
        V = a * log(U1/(1 - U1))
        Y = alpha * exp(V)
        Z = U1**2 * U2
        W = b + q*V - Y

        # Paso 3: pretest
        if W + d - theta*Z >= 0:
            samples.append(Y)
            continue

        # Paso 4: test completo
        if W >= log(Z):
            samples.append(Y)

    return np.array(samples)

# Función general para generar Gamma(α,β)

def generar_gamma(alpha, beta=1, n_samples=1):
    """
    Generador general de variables Gamma(α,β) que selecciona el algoritmo apropiado.

    Parámetros:
        alpha: parámetro de forma (alpha > 0)
        beta: parámetro de escala (beta > 0)
        n_samples: número de muestras a generar

    Devuelve:
        array de muestras Gamma(α,β)
    """
    if alpha <= 0 or beta <= 0:
        raise ValueError("alpha y beta deben ser positivos")

    if alpha == 1:
        # Caso exponencial
        samples = -np.log(np.random.uniform(size=n_samples)) * beta
    elif alpha < 1:
        # Algoritmo de Ahrens-Dieter para 0 < α < 1
        samples = gamma_ahrens_dieter(alpha, n_samples) * beta
    else:
        # Algoritmo de Cheng para α > 1
        samples = gamma_cheng(alpha, n_samples) * beta

    return samples

# Ejemplo de uso y visualización

# Parámetros
alpha = 0.7  # Prueba con valores <1 y >1
beta = 2.0
n_samples = 100000

# Generar muestras
samples = generar_gamma(alpha, beta, n_samples)

# Visualización
plt.figure(figsize=(10, 6))
counts, bins, _ = plt.hist(samples, bins=100, density=True,
                          alpha=0.7, edgecolor='black',
                          label='Datos simulados')

# PDF teórica para comparación
x_teorico = np.linspace(0, np.max(samples), 1000)
pdf_teorica = gamma_dist.pdf(x_teorico, a=alpha, scale=beta)
plt.plot(x_teorico, pdf_teorica, 'r-', lw=2, label='PDF teórica')

plt.title(f'Distribución Gamma (α={alpha}, β={beta})')
plt.xlabel('x')
plt.ylabel('Densidad de probabilidad')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()

# Estadísticas
media_teorica = alpha * beta
varianza_teorica = alpha * beta**2

print("\n--- Resultados teóricos ---")
print(f"Media teórica: {media_teorica:.4f}")
print(f"Varianza teórica: {varianza_teorica:.4f}")
print(f"Desviación estándar teórica: {sqrt(varianza_teorica):.4f}")

print("\n--- Resultados empíricos ---")
print(f"Media muestral: {np.mean(samples):.4f}")
print(f"Varianza muestral: {np.var(samples):.4f}")
print(f"Desviación estándar muestral: {np.std(samples):.4f}")
```

**Weibull** $X \sim We(k,\lambda)$

Función de densidad
\begin{equation}
f(x)=
\begin{cases}
\lambda \alpha (\lambda x)^{\alpha-1}e^{-(\lambda x)\alpha}  & \text{si } x > 0\\
0 & \text{si } x \leq 0
\end{cases}
\end{equation}

image.png

Funcion Acumulativa
\begin{equation}
F(x)=
\begin{cases}
1 - e^{-(\lambda x)\alpha}  & \text{si } x > 0\\
0 & \text{si } x \leq 0
\end{cases}
\end{equation}

image.png

Media

\begin{equation}
\mu =  \frac{1}{\lambda} \, \Gamma\left(1 + \frac{1}{\alpha}\right)
\end{equation}

Varianza
\begin{equation}
\sigma^2 =  \frac{1}{\lambda^2} \left[ \Gamma \left( 1 + \frac{2}{\alpha} \right) - \Gamma^2 \left( 1 + \frac{1}{\alpha} \right) \right]
\end{equation}

Desviacion estandar
\begin{equation}
\sigma =  \sqrt{\frac{1}{\lambda^2} \left[ \Gamma \left( 1 + \frac{2}{\alpha} \right) - \Gamma^2 \left( 1 + \frac{1}{\alpha} \right) \right]}
\end{equation}

**Pseudocidigo**

1.   Generar $U \sim U(0,1)$
2.   Definir $$ x = \beta -In(U)^{1/\alpha} $$
3.   Retornar las x`s



```python
import numpy as np
import matplotlib.pyplot as plt
from math import gamma

def generar_weibull(alpha, beta, n):
    U = np.random.uniform(0, 1, n)        #Genera n números aleatorios U(0,1)
    x = beta * (-np.log(U)) ** (1/alpha)  #Transformada inversa Weibull
    return x

#Parámetros de la distribución Weibull
alpha = 2.0  #Parámetro de forma (k)
beta = 1.5    #Parámetro de escala (λ)
n = 100000    #Número de muestras

#Generar datos
datos = generar_weibull(alpha, beta, n)

#Grafica
plt.figure(figsize=(10, 6))
plt.hist(datos, bins=100, density=True, alpha=0.7, label='Datos simulados')

#PDF teórica para comparación
x_teorico = np.linspace(0, np.max(datos), 1000)
pdf_teorica = (alpha/beta) * (x_teorico/beta)**(alpha-1) * np.exp(-(x_teorico/beta)**alpha)

plt.title(f"Distribución Weibull")
plt.xlabel("x")
plt.ylabel("Densidad de probabilidad")
plt.show()

#Cálculos teóricos y empíricos
media_teorica = beta * gamma(1 + 1/alpha)
varianza_teorica = beta**2 * (gamma(1 + 2/alpha) - (gamma(1 + 1/alpha))**2)

print("\n--- Resultados teóricos ---")
print(f"Media teórica: {media_teorica:.4f}")
print(f"Varianza teórica: {varianza_teorica:.4f}")
print(f"Desviación estándar teórica: {np.sqrt(varianza_teorica):.4f}")

print("\n--- Resultados empíricos ---")
print(f"Media muestral: {np.mean(datos):.4f}")
print(f"Varianza muestral: {np.var(datos):.4f}")
print(f"Desviación estándar muestral: {np.std(datos):.4f}")
```

Media

\sigma^2 =  \frac{\beta^2}{(\alpha-1)^2(\alpha-2)}

\sigma =  \frac{\beta}{(\alpha-1)(\alpha-2)^2}

**Pseudocodigo**

1. Generar $ U \sim U(0,1) $\\
2. Calcular $ y = \alpha(1 - U, \text{forma} = \alpha, \text{tasa} = \beta)$ \\
3. Retornar $ x = \frac{1}{y}$






```python
#Librerías
import numpy as np
import matplotlib.pyplot as plt
from random import random
from math import gamma, log

def gamma_inversa(alpha, n):
    from scipy.stats import gamma as gamma_dist
    return gamma_dist.rvs(a=alpha, scale=1.0, size=n)

def pearson_type_v(alpha, beta, n):
    Y = gamma_inversa(alpha, n)
    X = beta / Y
    return X

#Parámetros
alpha = 3.0
beta = 2.0
n = 10000

#Simular datos
A = pearson_type_v(alpha, beta, n)

# Histograma
plt.hist(A, bins=100, density=True, alpha=0.7, color='plum')
plt.title(f"Distribución Pearson Type V (Inversa Gamma)")
plt.xlabel("Valor")
plt.ylabel("Densidad")
plt.grid(True)
plt.show()

#Estadísticas teóricas (válidas si α > 1 o α > 2)
media_teorica = beta / (alpha - 1) if alpha > 1 else np.nan
varianza_teorica = (beta ** 2) / ((alpha - 1) ** 2 * (alpha - 2)) if alpha > 2 else np.nan
desviacion_teorica = np.sqrt(varianza_teorica) if alpha > 2 else np.nan


print("\nResultados teóricos")
print(f"Media teórica: {media_teorica:.4f}" if not np.isnan(media_teorica) else "Media teórica: No existe")
print(f"Varianza teórica: {varianza_teorica:.4f}" if not np.isnan(varianza_teorica) else "Varianza teórica: No existe")
print(f"Desviación estándar teórica: {desviacion_teorica:.4f}" if not np.isnan(desviacion_teorica) else "Desviación estándar teórica: No existe")

print("\nResultados numéricos")
print(f"Media muestral: {np.mean(A):.4f}")
print(f"Varianza muestral: {np.var(A):.4f}")
print(f"Desviación estándar muestral: {np.std(A):.4f}")
```

**Pearson type VI**  $ PT6(\alpha_1, \alpha_2, \beta) $


Funcion de densidad

\begin{equation}
f(x) =
\begin{cases}
\frac{(x/\beta)^{\alpha_1-1}}{\beta B (\alpha_1, \alpha_2)[1+(x / \beta)]^{\alpha_1 + \alpha_2} } & \text{ si }  x > 0 \\
0 &   \text{otro caso}
\end{cases}
\end{equation}

funcion acumulativa
\begin{equation}
F(x)=
\begin{cases}
F_B(\frac{x}{x + \beta})  & \text{ si } x > 0 \\
0 &   \text{otro caso}
\end{cases}
\end{equation}


Captura de pantalla 2025-05-23 152740.png






Media

\begin{equation}
\mu= \frac{\beta \alpha_1}{\alpha_2 -1}
\end{equation}

Varianza
\begin{equation}
\sigma^2 =  \frac{\beta^2 \alpha_1 (\alpha_1 + \alpha_2 - 1)}{(\alpha_2 - 1)^2(\alpha_2 - 2)}
\end{equation}

Desviacion estandar
\begin{equation}
\sigma =  \frac{ \beta \sqrt { \alpha_1 (\alpha_1 + \alpha_2 - 1)}}{(\alpha_2 - 1)\sqrt{\alpha_2 - 2}}
\end{equation}


**Pseudocidigo**

\begin{align*}
1. &\ \text{Generar } U \sim U(0,1) \\
2. &\ \text{Calcular } z = \beta(U, \alpha_1, \alpha_2) \\
3. &\ \text{Calcular } x = \frac{z}{\beta (1 - z)} \\
4. &\ \text{Retornar } x
\end{align*}





```python
#Librerías
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta as beta_dist

def pearson_type_vi(alpha, beta, n):
    U = beta_dist.rvs(a=alpha, b=beta, size=n)
    X = U / (1 - U)
    return X

#Parámetros
alpha = 2.5
beta = 3.0
n = 10000

#Simular datos
A = pearson_type_vi(alpha, beta, n)

#Histograma
plt.hist(A, bins=100, density=True, alpha=0.7, color='orange')
plt.title(f"Distribución Pearson Type VI")
plt.xlabel("Valor")
plt.ylabel("Densidad")
plt.grid(True)
plt.show()


media_teorica = alpha / (beta - 1) if beta > 1 else np.nan
varianza_teorica = (alpha * (alpha + beta - 1)) / ((beta - 2) * (beta - 1) ** 2) if beta > 2 else np.nan
desviacion_teorica = np.sqrt(varianza_teorica) if beta > 2 else np.nan


print("\nResultados teóricos")
print(f"Media teórica: {media_teorica:.4f}" if not np.isnan(media_teorica) else "Media teórica: No existe")
print(f"Varianza teórica: {varianza_teorica:.4f}" if not np.isnan(varianza_teorica) else "Varianza teórica: No existe")
print(f"Desviación estándar teórica: {desviacion_teorica:.4f}" if not np.isnan(desviacion_teorica) else "Desviación estándar teórica: No existe")

print("\nResultados numéricos")
print(f"Media muestral: {np.mean(A):.4f}")
print(f"Varianza muestral: {np.var(A):.4f}")
print(f"Desviación estándar muestral: {np.std(A):.4f}")
```

##**Log-logística** $LL( \alpha, \beta )$


Funcion de densidad

\begin{equation}
f(x) =
\begin{cases}
\frac{\alpha (x / \beta )^{\alpha -1}}{\beta [1+ (x /\beta)^{-\alpha}]} &  \text{ si } x > 0 \\
0 & \text{otro caso}
\end{cases}
\end{equation}

funcion acumulativa
\begin{equation}
F(x)=
\begin{cases}
\frac{1}{1+ (x /\beta)^{-\alpha}}  & \text{ si } x > 0 \\
0 & \text{ otro caso  }
\end{cases}
\end{equation}

Media

\begin{equation}
\mu= \beta \theta \text{ cosecant}(\theta)
\end{equation}

Varianza
\begin{equation}
\sigma^2 =  \beta^2\theta[2\text{cosecant}(2\theta) - \theta [\text{cosecant}(\theta)]^2]
\end{equation}

Desviacion estandar
\begin{equation}
\sigma =  \beta \theta ^2[2\text{cosecant}(2\theta) - \theta [\text{cosecant}(\theta)]]
\end{equation}



***Pseudocodigo**


1. Generar $ U \sim U(0,1)$ \\
2. Calcular $ x = \beta \left( \frac{U}{1 - U} \right)^{1/\alpha}$ \\
3. Retornar  $x$





```python
#Librerías
import numpy as np
import matplotlib.pyplot as plt
from random import random


def loglogistica(alpha, beta, n):
    x = []
    for _ in range(n):
        u = random()
        valor = alpha * (u / (1 - u)) ** (1 / beta)  # Transformada inversa
        x.append(valor)
    return x

#Parámetros
alpha = 2.0
beta = 3.0
n = 10000


A = loglogistica(alpha, beta, n)

#Histograma
plt.hist(A, bins=100, density=True, alpha=0.7, color='skyblue')
plt.title(f"Distribución Log-logística LL(α={alpha}, β={beta})")
plt.xlabel("Valor")
plt.ylabel("Densidad")
plt.grid(True)
plt.show()


if beta > 1:
    media_teorica = (np.pi * alpha / beta) / np.sin(np.pi / beta)
else:
    media_teorica = np.nan  # No existe

if beta > 2:
    varianza_teorica = (alpha ** 2) * (2 * np.pi / beta) / np.sin(2 * np.pi / beta) - media_teorica ** 2
    desviacion_teorica = np.sqrt(varianza_teorica)
else:
    varianza_teorica = np.nan
    desviacion_teorica = np.nan


print("\nResultados teóricos")
print(f"Media teórica: {media_teorica:.4f}" if not np.isnan(media_teorica) else "Media teórica: No existe")
print(f"Varianza teórica: {varianza_teorica:.4f}" if not np.isnan(varianza_teorica) else "Varianza teórica: No existe")
print(f"Desviación estándar teórica: {desviacion_teorica:.4f}" if not np.isnan(desviacion_teorica) else "Desviación estándar teórica: No existe")

print("\nResultados numéricos")
print(f"Media muestral: {np.mean(A):.4f}")
print(f"Varianza muestral: {np.var(A):.4f}")
print(f"Desviación estándar muestral: {np.std(A):.4f}")
```

f(x) =

0 & \text{otro caso}
\end{cases}
\end{equation}

**Pseudocidigo**


1. Generar $ Z \sim N(0,1)$ \\
2. Calcular $ x = \xi + \lambda \cdot \sinh\left(\frac{Z - \gamma}{\delta} \right) $\\
3. Retornar $ x $


```python
#Librerías
import numpy as np
import matplotlib.pyplot as plt

def johnson_su_simulacion(gamma, delta, xi, lambd, n):
    Z = np.random.normal(0, 1, n)  # Z ~ N(0, 1)
    X = xi + lambd * np.sinh((Z - gamma) / delta)
    return X

#Parámetros
gamma = 0.0
delta = 1.0
xi = 0.0
lambd = 1.0
n = 10000

#Simulacion
A = johnson_su_simulacion(gamma, delta, xi, lambd, n)

#Histograma
plt.hist(A, bins=100, density=True, alpha=0.7, color='mediumseagreen')
plt.title("Distribución Johnson SU (simulada desde normal estándar)")
plt.xlabel("Valor")
plt.ylabel("Densidad")
plt.grid(True)
plt.show()


print("\nResultados numéricos")
print(f"Media muestral: {np.mean(A):.4f}")
print(f"Varianza muestral: {np.var(A):.4f}")
print(f"Desviación estándar muestral: {np.std(A):.4f}")
```

**Uniforme** $X \sim U(a,a+1,...,b) $


Funcion de densidad

\begin{equation}
f(x) =
\begin{cases}
\frac{1}{b-a+1} &  \text{ si } x \in (a,a+1,...,b) \\
0 & \text{otro caso}
\end{cases}
\end{equation}


image.png


funcion acumulativa
\begin{equation}
F(x)=
\begin{cases}
0  & \text{ si } x < a \\
\frac{|x|-a+1}{b-a+1}  & \text{ si } a \leq x < b; \\
1 & \text{ si } x \geq b
\end{cases}
\end{equation}


image.png




Media

\begin{equation}
\mu= \frac{a+b}{2}
\end{equation}

Varianza
\begin{equation}
\sigma^2 =  \frac{(b-a+1)^2-1}{12}
\end{equation}

Desviacion estandar
\begin{equation}
\sigma =  \sqrt{\frac{(b-a+1)^2-1}{12}}
\end{equation}


**Pseudocidigo**

1. Generar $ U \sim U(0,1)$ \\
2. Calcular $ x = \lfloor a + (b - a + 1) \cdot U \rfloor$ \\
3. Retornar  $x$




```python
import numpy as np
import matplotlib.pyplot as plt
from random import random


def uniforme_discreta(n):
    a, b = 1, 6
    x = []
    for _ in range(n):
        U = random()
        val = a + int((b - a + 1) * U)
        x.append(val)
    return x

#Simulaciones
A = uniforme_discreta(10000)

#Histograma
plt.hist(A, bins=np.arange(0.5, 7.5, 1), density=True, edgecolor='black')
plt.title("Distribución Uniforme Discreta")
plt.xticks(range(1, 7))
plt.grid()
plt.show()


media = np.mean(A)
varianza = np.var(A)
desviacion = np.std(A)


print("Media:", round(media, 4))
print("Varianza:", round(varianza, 4))
print("Desviación estándar:", round(desviacion, 4))
```

**Bernoulli**  $ (p) $


Funcion de densidad

\begin{equation}
f(x) =
\begin{cases}
1 - p & \text{ si }  x = 0 \\
p &  \text{ si }  x = 1 \\
0 &   \text{otro caso}
\end{cases}
\end{equation}


image.png


funcion acumulativa
\begin{equation}
F(x)=
\begin{cases}
0  & \text{ si } x < 0 \\
1 - p  & \text{ si } 0 \leq x < 1; \\
1 & \text{ si } x \geq 1
\end{cases}
\end{equation}



image.png





Media

\begin{equation}
\mu= p
\end{equation}

Varianza
\begin{equation}
\sigma^2 =  p(1-p)
\end{equation}

Desviacion estandar
\begin{equation}
\sigma =  \sqrt{p(1-p)}
\end{equation}


**Pseudocidigo**


1. Generar $ U \sim U(0,1)$ \\
2. Si  $U < p$,  retornar $ x = 1$ \\
3. Si no, retornar $ x = 0 $






```python
import numpy as np
import matplotlib.pyplot as plt
from random import random

#Funcion
def bernoulli(p, n):
    x = []
    for _ in range(n):
        U = random()
        x.append(1 if U < p else 0)
    return x

#Simulaciones
A = bernoulli(0.3, 10000)

#Histograma
plt.hist(A, bins=[-0.5, 0.5, 1.5], density=True, edgecolor='black')
plt.title("Distribución Bernoulli (p=0.3)")
plt.xticks([0, 1])
plt.grid()
plt.show()

media = np.mean(A)
varianza = np.var(A)
desviacion = np.std(A)


print("Media:", round(media, 4))
print("Varianza:", round(varianza, 4))
print("Desviación estándar:", round(desviacion, 4))
```

f(x)=

p(1-p)^x &  \text{si } x \leq 0\\
0 & otro \;\; caso

\end{equation}

Funcion Acumulativa

F(x)=

1-(1-p)^{|x|+1}  & \text{si } x \geq 0\\
0 & \text{otro caso }

\end{equation}

image.png




Media

\begin{equation}
\mu= \frac{1-p}{p}
\end{equation}

Varianza
\begin{equation}
\sigma^2 =  \frac{1-p}{p^2}
\end{equation}

Desviacion estandar
\begin{equation}
\sigma =   \frac{\sqrt {1-p}} {p}
\end{equation}

**Pseudocidigo**

\begin{align*}
1. Generar $U \sim U(0,1) $\\
2. Calcular $ x = \left\lfloor \frac{\log(1 - U)}{\log(1 - p)} \right\rfloor$ \\
3. Retornar $x$





```python
import numpy as np
import matplotlib.pyplot as plt
from random import random
import math


def geometrica(p, n):
    x = []
    for _ in range(n):
        U = random()
        val = math.ceil(math.log(1 - U) / math.log(1 - p))
        x.append(val)
    return x

#Simulaciones
A = geometrica(0.3, 10000)

#Histograma
plt.hist(A, bins=np.arange(0.5, max(A)+1.5, 1), density=True, edgecolor='black')
plt.title("Distribución Geométrica (p=0.3)")
plt.grid()
plt.show()


media = np.mean(A)
varianza = np.var(A)
desviacion = np.std(A)


print("Media:", round(media, 4))
print("Varianza:", round(varianza, 4))
print("Desviación estándar:", round(desviacion, 4))
```

**Binomial negativa** $X \sim NegBin(r,p)$


Función de densidad
\begin{equation}
f(x)=
\begin{cases}
\binom{s + x - 1}{x} p^s (1 - p)^x & x = 0,1,2,\dots \\
0 & \text{otro caso}
\end{cases}
\end{equation}




image.png


Funcion Acumulativa
\begin{equation}
F(x)=
\begin{cases}
0  & \text{si } x < 0\\
\sum (\frac{s + i -1}{i})p^s(1-p)^i & \text{si } x \geq 0
\end{cases}
\end{equation}

Media

\begin{equation}
\mu= \frac{s(1-p)}{p}
\end{equation}

Varianza
\begin{equation}
\sigma^2 =  \frac{s(1-p)}{p^2}
\end{equation}

Desviacion estandar
\begin{equation}
\sigma =  \frac{\sqrt {s(1-p)}}{p}
\end{equation}

**Pseudocidigo**

1. Generar $ U \sim U(0,1)$ \\
2. Buscar el menor  $x $ tal que\\
   $\quad \sum_{k=0}^{x} \binom{k+r-1}{k} p^r (1-p)^k \geq U$ \\
3. Retornar  $ x $



```python
import numpy as np
import matplotlib.pyplot as plt
from random import random
import math


def binomial_negativa(r, p, n):
    x = []
    for _ in range(n):
        total = 0
        for _ in range(r):
            U = random()
            g = math.ceil(math.log(1 - U) / math.log(1 - p))
            total += g - 1
        x.append(total)
    return x

#Simulación
A = binomial_negativa(3, 0.3, 10000)

#Histograma
plt.hist(A, bins=np.arange(min(A) - 0.5, max(A) + 1.5, 1), density=True, edgecolor='black')
plt.title("Distribución Binomial Negativa (r=3, p=0.3)")
plt.grid()
plt.show()


media = np.mean(A)
varianza = np.var(A)
desviacion = np.std(A)

print("Media:", round(media, 4))
print("Varianza:", round(varianza, 4))
print("Desviación estándar:", round(desviacion, 4))
```

##**Poisson** $(\lambda)$

Función de densidad
\begin{equation}
f(x)=
\begin{cases}
\frac{e^{- \lambda } λ^x}{x !} &  \text{si } x \leq 0\\
0 & otro \;\; caso
\end{cases}
\end{equation}



image.png





Funcion Acumulativa
\begin{equation}
F(x)=
\begin{cases}
0  & \text{si } x < 0\\
e^{- \lambda} \sum \frac{\lambda ^i}{i!} & \text{si } x \geq 0
\end{cases}
\end{equation}

Media

\begin{equation}
\mu= \lambda
\end{equation}

Varianza
\begin{equation}
\sigma^2 =  \lambda
\end{equation}

Desviacion estandar
\begin{equation}
\sigma =  \sqrt{\lambda}
\end{equation}

**Pseudocidigo**


1. Generar $ U \sim U(0,1),\ F = e^{-\lambda},\ x = 0$ \\
2. Mientras $ F < U:$ \\
   $\quad x = x + 1 $\\
   $\quad F = F + \frac{e^{-\lambda} \lambda^x}{x!}$ \\
3. Retornar $ x$



```python
import numpy as np
import matplotlib.pyplot as plt
from random import random
import math

def poisson(lamb, n):
    x = []
    for _ in range(n):
        U = random()
        k = 0
        p = math.exp(-lamb)
        F = p
        while U > F:
            k += 1
            p *= lamb / k
            F += p
        x.append(k)
    return x

#Simulaciones
A = poisson(4, 10000)

#Histograma
plt.hist(A, bins=np.arange(min(A)-0.5, max(A)+1.5, 1), density=True, edgecolor='black')
plt.title("Distribución Poisson (λ = 4)")
plt.grid()
plt.show()


media = np.mean(A)
varianza = np.var(A)
desviacion = np.std(A)


print("Media:", round(media, 4))
print("Varianza:", round(varianza, 4))
print("Desviación estándar:", round(desviacion, 4))
```
