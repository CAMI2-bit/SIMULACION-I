```python
import numpy as np
import matplotlib.pyplot as plt
from random import random

def inversa(n):
  x = []
  for i in range(n):
    R = random()
    if 0<=R<=0.5:
      x.append((2*R)**(1/2))
    elif 0.5<R<=1:
      x. append(2*R)
  return x

a = inversa(10000)

plt.hist(a, bins=100)

print(np.mean(a))
print(np.var(a))
```
