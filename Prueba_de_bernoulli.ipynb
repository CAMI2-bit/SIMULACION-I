```python

import numpy as np
import matplotlib.pyplot as plt

def F_bar(u, p):
    return (u > 1 - p).astype(int)

def simular_bernoulli_con_F_bar(p, n_samples=1000):
    """
    Simula variables Bernoulli(p) usando F̅(u)
    """
    u = np.random.uniform(0, 1, n_samples)
    x = F_bar(u, p)
    return x

# Parámetros
p = 0.2
n_samples = 1000

# Simulación
muestras = simular_bernoulli_con_F_bar(p, n_samples)

# Histograma
plt.figure(figsize=(6, 4))
plt.hist(muestras, bins=[-0.5, 0.5, 1.5], edgecolor='black', rwidth=0.8)
plt.xticks([0, 1])
plt.xlabel('Valor de X')
plt.ylabel('Frecuencia')
plt.title(f'Histograma de muestras Bernoulli(p={p}) usando F̅(u)')
plt.grid(True)
plt.show()
```

```python

import numpy as np
import matplotlib.pyplot as plt

def F_bar(u, p):

    if u > 1 - p:
        return 1
    else:
        return 0

def simular_bernoulli_con_F_bar(p, n_samples=1000):

    muestras = []
    for _ in range(n_samples):
        u = np.random.uniform(0, 1)
        x = F_bar(u, p)
        muestras.append(x)
    return np.array(muestras)

# Parámetros
p = 0.2
n_samples = 1000

# Simulación
muestras = simular_bernoulli_con_F_bar(p, n_samples)

# Estadísticas
media = np.mean(muestras)
varianza = np.var(muestras)

# Mostrar estadísticas
print(f"Media muestral: {media:.4f}")
print(f"Varianza muestral: {varianza:.4f}")

# Histograma
plt.figure(figsize=(6, 4))
plt.hist(muestras, bins=[-0.5, 0.5, 1.5], edgecolor='black', rwidth=0.8)
plt.xticks([0, 1])
plt.xlabel('Valor de X')
plt.ylabel('Frecuencia')
plt.title(f'Histograma de muestras Bernoulli(p={p})')
plt.grid(True)
plt.show()

# Gráfica de la función F̅(u)
u_vals = np.linspace(0, 1, 1000)
f_vals = [F_bar(u, p) for u in u_vals]

plt.figure(figsize=(6, 4))
plt.step(u_vals, f_vals, where='post', color='blue', label=r'$\overline{F}(u)$')
plt.axvline(1 - p, color='red', linestyle='--', label=r'$u = 1 - p$')
plt.xlabel('u')
plt.ylabel(r'$\overline{F}(u)$')
plt.title(r'Función $\overline{F}(u)$ para Bernoulli(p)')
plt.ylim(-0.1, 1.1)
plt.legend()
plt.grid(True)
plt.show()
```
