```python
from sympy import fourier_series, pi, plot, Piecewise, sin, symbols, integrate, exp
from sympy.abc import x, n
```

```python
# Definir la función en el intervalo [0,1]
f = Piecewise((exp(2*x), (x >= 0) & (x <= 1)))
```

```python
# Coeficientes de la serie de Fourier en senos
L = 1  # Periodo simétrico [-1,1]
b_n = (2/L) * integrate(exp(2*x) * sin(n * pi * x / L), (x, 0, L))
```

```python
# Construir la serie de Fourier en senos hasta un cierto número de términos
N = 5  # Número de términos
serie_senos = sum(b_n.subs(n, k) * sin(k * pi * x / L) for k in range(1, N+1))
```

```python
serie_senos
```

```python
# Graficar la función y su aproximación de Fourier
p = plot(f, serie_senos, (x, -1, 1), show=False, legend=True)
p[0].label = 'f(x)'
p[1].label = f'Serie de Fourier (n={N})'
p.show()
```

```python
from sympy import fourier_series, pi, plot, Piecewise, cos, symbols, integrate, exp
from sympy.abc import x, n
```

```python
# Definir la función en el intervalo [0,1]
f = Piecewise((exp(2*x), (x >= 0) & (x <= 1)))
```

```python
# Coeficientes de la serie de Fourier en cosenos
L = 1  # Periodo simétrico [-1,1]
a_n = (2/L) * integrate(exp(2*x) * cos(n * pi * x / L), (x, 0, L))
```

```python
# Construir la serie de Fourier en cosenos hasta un cierto número de términos
N = 5  # Número de términos
serie_cosenos = sum(a_n.subs(n, k) * cos(k * pi * x / L) for k in range(1, N+1))
```

```python
serie_cosenos
```

```python
# Graficar la función y su aproximación de Fourier
p = plot(f, serie_cosenos, (x, -1, 1), show=False, legend=True)
p[0].label = 'f(x)'
p[1].label = f'Serie de Fourier (n={N})'
p.show()
```
