```python
import numpy as np

def laplace(n):
    lista = []
    for _ in range(n):
        R1 = np.random.uniform(0, 1)
        R2 = np.random.uniform(0, 1)
        if R1 <= 0.5:
            lista.append(np.log(R2))
        else:
            lista.append(-np.log(R2))
    return lista


muestra = laplace(100000)

plt.hist(muestra,bins=50, density=True )
plt.show()
```
